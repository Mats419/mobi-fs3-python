**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img, size): #img ist das Bild, size ist die filter size determining the filter neigh.
    result = numpy.zeros(img.shape) #man erstellt ein leeres bild, in welchem die neuen pixel def. werden
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape): #man durchlÃ¤uft jedes Pixel
        neighborhood_sum = 0
        if 511-n>=p[0]>=n:
            if 511-n>=p[1]>=n:
                for q in numpy.ndindex(size, size): 
                    x = p[0]+q[0]-n
                    y = p[1]+q[1]-n
                    neighborhood_sum = neighborhood_sum + img[x,y]
                result[p] = neighborhood_sum / size*size
    # write your code here
    return result
```
