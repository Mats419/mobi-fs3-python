**<a id="#">Task 4.1.1.</a>** Load the image `data/astronaut.png` and show it:

**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img):
    result = numpy.zeros(img.shape)
    n = 1
    mask = numpy.asarray([[-1,0,1],[-1,0,1],[-1,0,1]])
    for p in numpy.ndindex(img.shape):
        neighborhood_sum = 0
        if img.shape[0]-n>p[0]>=n:
            if img.shape[1]-n>p[1]>=n:
                for q in numpy.ndindex(3, 3): 
                    m = mask[q]
                    x = p[0]+q[0]-n
                    y = p[1]+q[1]-n
                    neighborhood_sum = neighborhood_sum + img[x,y]*m
                    z = 1/6
                result[p] = neighborhood_sum*z
    return result
    
def prewitt_v(img):
    result = numpy.zeros(img.shape)
    n = 1
    mask = numpy.asarray([[-1,-1,-1],[0,0,0],[1,1,1]])
    for p in numpy.ndindex(img.shape):
        neighborhood_sum = 0
        if img.shape[0]-n>p[0]>=n:
            if img.shape[1]-n>p[1]>=n:
                for q in numpy.ndindex(3, 3): 
                    m = mask[q]
                    x = p[0]+q[0]-n
                    y = p[1]+q[1]-n
                    neighborhood_sum = neighborhood_sum + img[x,y]*m
                    z = 1/6
                result[p] = neighborhood_sum*z
    return result
```

```python
testh = prewitt_h(img)
# plt.imshow(testh,'gray')

testv = prewitt_v(img)
# plt.imshow(testv,'gray')

plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.imshow(testh,'gray')
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(testv, 'gray')
plt.colorbar()
```
